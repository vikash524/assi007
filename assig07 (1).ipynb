```python
#Q1. What is Abstraction in OOPs? Explain with an example.


# Abstraction in OOPs refers to the concept of hiding the internal implementation details and showing only the necessary features of an object to the outside world.

# Example:
from abc import ABC, abstractmethod

class Vehicle(ABC):
    def __init__(self, name):
        self.name = name
    
    @abstractmethod
    def drive(self):
        pass

class Car(Vehicle):
    def drive(self):
        print(f"{self.name} is being driven on the road.")

# Creating an instance of Car
car = Car("Toyota")
car.drive()
#Q2. Differentiate between Abstraction and Encapsulation. Explain with an example.


# Abstraction focuses on showing only the necessary features of an object, hiding the internal implementation details.
# Encapsulation is the bundling of data and methods that operate on the data into a single unit (class), and restricting access to some of the object's components.

# Example of Abstraction:
class Animal:
    def speak(self):
        pass

class Dog(Animal):
    def speak(self):
        print("Dog barks")

# Example of Encapsulation:
class Car:
    def __init__(self, make, model):
        self.make = make
        self.model = model
    
    def drive(self):
        print(f"Driving {self.make} {self.model}")

# Accessing Car class encapsulated attributes and method
car = Car("Toyota", "Camry")
car.drive()
#Q3. What is abc module in Python? Why is it used?


# The abc (Abstract Base Classes) module in Python provides tools for creating abstract base classes. 
# Abstract base classes are classes that are meant to be subclassed but not instantiated on their own. 
# They define methods that must be implemented by their subclasses.

# Example:
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Rectangle(Shape):
    def __init__(self, length, breadth):
        self.length = length
        self.breadth = breadth
    
    def area(self):
        return self.length * self.breadth

# Creating an instance of Rectangle
rect = Rectangle(5, 4)
print("Area of Rectangle:", rect.area())
#Q4. How can we achieve data abstraction?

# Data abstraction can be achieved in Python by using abstract classes and methods. 
# Abstract classes define methods without implementation and can have abstract methods that must be implemented by their subclasses.

# Example:
from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

class Dog(Animal):
    def sound(self):
        print("Dog barks")

# Creating an instance of Dog
dog = Dog()
dog.sound()
#Q5. Can we create an instance of an abstract class? Explain your answer.

# No, we cannot create an instance of an abstract class directly. 
# Abstract classes are meant to be subclassed, and the subclasses must provide implementations for the abstract methods defined in the abstract class.

# Example:
from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

# Attempting to create an instance of the abstract class Animal will raise an error
  # This will raise an error
```
